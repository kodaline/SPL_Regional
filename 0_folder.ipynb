```python
from IPython.core.display import display, HTML
import os

def list_files(startpath):
    output = "<ul>"
    
    for root, dirs, files in os.walk(startpath):
        # Remove hidden directories from the 'dirs' list in-place
        dirs[:] = [d for d in dirs if not d.startswith('.')]
        
        level = root.replace(startpath, '').count(os.sep)
        indent = '    ' * level
        
        # Skip hidden folders (those starting with .)
        if os.path.basename(root).startswith('.'):
            continue
            
        output += f"{indent}<li><strong>{os.path.basename(root)}/</strong></li><ul>"
        
        # Sort the files alphabetically and limit to top 10
        files = sorted(files)[:10]
        
        subindent = '    ' * (level + 1)
        for f in files:
            output += f"{subindent}<li>{f}</li>"
        
        output += f"{indent}</ul>"
        
    output += "</ul>"
    display(HTML(output))

# Replace '.' with the path to the directory you want to print.
list_files('.')
```
